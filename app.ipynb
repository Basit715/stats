```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
```

```python
dice = [1,2,3,4,5,6,5,4,3,2,1]
sns.histplot(dice, kde=True)
```

```python
# To check whether our data is normally distributed or not by q-q plot
# if in the q-q plot the real throritical data and real data follow a straight line then our data is normally distributed otherwise data is not normally distributed
def polt_data(sample):
     plt.figure(figsize=(10,6))
     plt.subplot(1,2,1)
     sns.histplot(sample, kde=True)     #our real data
     plt.subplot(1,2,2)
     stat.probplot(sample, dist='norm', plot=pylab)  # used to assume that the data we have is normally distributed
     plt.show()
```

```python
polt_data(real_log_normal)
```

```python
polt_data(rand)
```
